```python
# Importing the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

# Reading Dataset

```python
Data=pd.read_csv('Iris.csv')
Data.head(10)
```

# Data Visualization

```python
plt.scatter(Data["SepalLengthCm"],Data["Species"],color="dodgerblue",marker='*')
plt.xlabel("SepalLengthCm")
plt.ylabel("Species")
plt.title("SepalLengthCm VS Species")
plt.show()
```

```python
sns.catplot(x="SepalLengthCm",y="PetalWidthCm",data=Data,kind='swarm',col="Species")
```

```python
sns.scatterplot(x="SepalLengthCm",y="PetalLengthCm",data=Data,hue="Species")
```

# Model Building & Training

```python
x=Data.drop(columns='Species',axis=True)
y=Data.Species
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
Clf = DecisionTreeClassifier()
temp=Clf.fit(X_train, y_train)
predictions = Clf.predict(X_test)
print('Accuracy Score:',Clf.score(X_test,y_test)*100)
```

# Decision Tree of classified Data

```python
plt.figure(figsize=(10,10))
plot_tree(temp,rounded=True,filled=True)
plt.show()
```
